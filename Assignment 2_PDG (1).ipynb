```python
# Coin Flip, 10,000 flips, store in a list, identify no. of streaks (5+)

import random

number_of_coins = 10000
heads_tails = []
heads = 0
tails = 0

for i in range(number_of_coins):
    number_of_coins = random.randint(0, 1)
    if number_of_coins == 0:
        heads_tails.append("T")
        tails += 1
    else:
        heads_tails.append("H")
        heads += 1

count_streaks=0
count_sequence = 0
for i in range(number_of_coins):
    if heads_tails[i]==heads_tails[i-1]:
        count_sequence +=1
    else:
        count_sequence = 0
    if count_sequence==4:
        count_streaks +=1 
        count_sequence =0
        
print(heads_tails)
print("Number of heads: ", heads)
print("Number of tails: ", tails)
print("Number of streaks: ", count_streaks)
   
```
